# apply,map

```python
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv',index_col=0)
Ans =df.shape

```

```python
df
```

```python
# 57번
# 컬럼에 map 적용하기
# map(lambda x: x)
dic = {"Unknown" : "N",
    "Less than $40K" : "a",
"$40K - $60K" : "b",
"$60K - $80K" : "c",
"$80K - $120K" : "d",
"$120K +" : "e"}
df["newincome"] = df.Income_Category.map(lambda x: dic[x])
df["newincome"]
```

```python
# 58번
# 컬럼에 apply 적용하기
# apply는 이전에 만든 함수를 사용
def change_category(x):
    if x=="Unknown":
        return "N"
    elif x=="Less than $40K":
        return "a"
    elif x=="$40K - $60K":
        return "b"
    elif x=="$60K - $80K":
        return "c"
    elif x=="$80K - $120K":
        return "d"
    elif x=="$120K +":
        return "e"
    
df["newIncome"] = df.Income_Category.apply(change_category)

df["newIncome"]
```

```python
df["Customer_Age"].max()
```

```python
def category(x):
    if 0<=x<=9:
        return 0
    elif 10<=x<=19:
        return 10
    elif 20<=x<=29:
        return 20
    elif 30<=x<=39:
        return 30
    elif 40<=x<=49:
        return 40
    elif 50<=x<=59:
        return 50
    elif 60<=x<=69:
        return 60
    elif 70<=x<=79:
        return 70

df["AgeState"] = df.Customer_Age.apply(category)
df.groupby("AgeState").size()
```

```python
# 59번
# 하나의 컬럼의 행별 개수를 구하려면 value_counts를 사용하는 것도 방법 중의 하나!
df["AgeState"] = df.Customer_Age.map(lambda x : x//10*10)
df["AgeState"].value_counts().sort_index()
```

```python
df["newEduLevel"] = df.Education_Level.map(lambda x: 1 if "Graduate" in x else 0)
df["newEduLevel"].value_counts(0)
```

```python
df["newLimit"] = df.Credit_Limit.map(lambda x : 1 if x>=4500 else 0)
df["newLimit"].value_counts()
```

```python
# 62번
# 컬럼 두개 이상을 사용하여 조건이 주어질 때 apply 활용
# apply를 적용할 축을 줘야한다.
def newState(x):
    if x.Marital_Status=="Married" and x.Card_Category=="Platinum":
        return 1
    else:
        return 0
df["newState"] = df.apply(newState,axis=1)
df["newState"].value_counts()
```

```python
df.loc[:,"Gender"] = df.Gender.map(lambda x: "male" if x=="M" else "female")
df["Gender"].value_counts()
```

```python
import pandas as pd
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv'
df = pd.read_csv(DataUrl)
```

```python
df.info()
```

```python
# astype 또는 pd.to_datetime을 사용
df["Yr_Mo_Dy"] = df["Yr_Mo_Dy"].astype("datetime64")
df["Yr_Mo_Dy"] = pd.to_datetime(df["Yr_Mo_Dy"])
df["Yr_Mo_Dy"].head()
```

```python
# 66번
# datetime 형태면 dt.year로 년도만 뽑을 수 있다.
# dt는 시리즈형태를 datetime으로 바꿔주는 것이다.
df.Yr_Mo_Dy.unique()
df.Yr_Mo_Dy.dt.year.unique()
```

```python
# 67번
# datetime.date
def fix_century(x):
    import datetime
    
    if x.year>=2061:
        year = x.year-100
    else:
        year = x.year
        
    return pd.to_datetime(datetime.date(year, x.month, x.day))
df.Yr_Mo_Dy.apply(fix_century)
```

```python
# 68
# groupby의 기준은 시리즈형태다.
df.groupby(df.Yr_Mo_Dy.dt.year).mean()
```

```python
# 69
# weekday는 요일별로 숫자로 매핑한다.
# week는 일년을 단위로 하여 숫자로 매핑한다.
df.Yr_Mo_Dy.dt.weekday
```

```python
df.Yr_Mo_Dy.dt.weekday.map(lambda x: 1 if x==5 or x==6 else 0)
```

```python
# 71
# 각 달의 평균 구하기
df.groupby(df.Yr_Mo_Dy.dt.month).mean()
```
