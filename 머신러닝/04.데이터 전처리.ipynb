# 데이터 인코딩

# 레이블 인코딩(Label Encoding)

```python
# target에만 적용해야한다
from sklearn.preprocessing import LabelEncoder

items= ["jean","skirt","jean","onepiece","jean","skirt","skirt","jean","onepiece"]

lbl_encoder = LabelEncoder()
lbl_encoder.fit(items)
```

```python
# 문자열을 숫자로 변환(Transform)하는 과정이다
labels = lbl_encoder.transform(items)
labels
```

```python
# 숫자를 다시 문자열로 돌려주는 과정을 디코디(Decode)
lbl_encoder.inverse_transform([2,1,0,0,1,2,0,0,1])
```

```python
# 클래스 확인
lbl_encoder.classes_
```

# 원-핫 인코딩(One-Hot Encoding)

```python
from sklearn.preprocessing import OneHotEncoder

ohe_encoder = OneHotEncoder() # Feature(행렬로 표현되어야한다) , Label(Target) -> Vector 
```

```python
# ohe_encoder.fit(items) -> 에러
import numpy as np
# 1차원 배열을 2차원 배열로 바꿔주기
# vector -> matrix 화
# N -> N x 1
items_arr = np.array(items).reshape(-1,1)
items_arr
```

```python
# ohe_encoder.fit(items)
# ohe_labels = ohe_encoder.transform(items)

# fit과 transform을 한꺼번에
ohe_labels = ohe_encoder.fit_transform(items_arr)
ohe_labels
```

```python
print(ohe_labels) # 희소행렬(sparse matrix) , OHE -> 쓸모없는 항목이 많다. 메모리 소모가 많음
# 어느 위치에 1이 있는지 표현해준다. (행인덱스,열인덱스) . dummy
```

```python
# 압축된 희소행렬 형태를 풀어서 표현
ohe_labels.toarray()
```
