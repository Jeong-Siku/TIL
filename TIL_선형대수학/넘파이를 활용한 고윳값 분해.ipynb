```python
import numpy as np
```

```python
# np.linalg.eig -> 고윳값 분해 함수
A = np.array([[1, -2],
              [2, -3]])

l1, V1 = np.linalg.eig(A)

print("lambda : {}".format(l1))
print("V : {}".format(V1))
```

```python
B = np.array([[2, 3],
              [2, 1]])

l2, V2 = np.linalg.eig(B)

print("lambda : {}".format(l2))
print("V : {}".format(V2))
```

```python
C = np.array([[0, -1],
              [1, 0]])

l3, V3 = np.linalg.eig(C)

print("lambda : {}".format(l3))
print("V : {}".format(V3))
```

# 대각화

```python
B = np.array([[2,3],
             [2,1]])

lambda_b , V = np.linalg.eig(B)

lambda_b, V
```

```python
# 고윳값 행렬 만들기
L = np.diag(lambda_b)
L
```

```python
# 고유벡터행렬의 역행렬
V_inv = np.linalg.inv(V)
V_inv
```

```python
V @ L @ V_inv
```

# 대칭행렬을 랭크-1 행렬의 합으로 분해

```python
A = np.array([[60., 30., 20.],
              [30., 20., 15.],
              [20., 15., 12.]])
```

```python
w, V = np.linalg.eig(A)
```

```python
w # 고윳값 벡터
```

```python
V # 고유 벡터 행렬
```

```python
# 각각의 고유값 구하기
w1, w2, w3 = w

# 각각의 고유벡터 구하기
# 슬라이싱을 통해 차원 유지
v1, v2, v3 = V[:,0:1],V[:,1:2],V[:,2:3]

v1.shape, v2.shape, v3.shape
```

```python
A1 = v1 @ v1.T
A2 = v2 @ v2.T
A3 = v3 @ v3.T 

A1
```

```python
w1 * A1
# 1의 값이 대부분의 값을 가짐. w1이 누적분포의 대부분을을 차지하기 떄문에 이 값만 살펴도
# 노이즈제거 및 추천알고리즘에 쓰인다.
```

```python
w2 * A2
```

```python
w1 * A1 + w2 * A2
```

```python
w3 * A3
```

```python
w1 * A1 + w2 * A2 + w3 * A3
```

```python
# A의 역행렬
np.linalg.inv(A)
```

```python
# 고윳값의 역수를 구해서 다 더해주면 된다.
1/w1 * A1 + 1/w2 * A2 + 1/w3 * A3
```
