## 조건문 ( `Condition` )
- 특정한 조건을 만족하면 ( `True` ) 코드를 실행
- `if`, `elif`, `else`
  * `if` : `만약에 ~라면 ~ 해라`
  * `elif (else if의 약어)` : `그게 아니고 ~라면 ~ 해라`
  * `else` : 그게 아니면 `~ 해라`

### 조건문의 구조
```python
if [condition]:
  code context
elif [condition]:
  code context
elif [condition]:
  code context
...
else:
  code context
```
* 조건 부분에 bool 타입 데이터(True / False)가 온다.



### Code Context
- Context : 코드 문맥
- 특정한 상황에 실행해야 할 코드 뭉터기
- 파이썬에서는 Context를 `indent(들여쓰기)`로 구분
- `indent`는 공백 2칸이나 4칸으로 설정을 하는 것이 관례
- 처음에 `indent`를 2칸으로 지정했으면 계속 2칸으로 쓰고, 4칸으로 지정했으면 계속 4칸으로 쓰면 된다.


````python
a = 3
b = 2

if a > b :     # 만약에 a가 b보다 작다면
  print("Hi")  # "Hi"를 출력
  print("Bye") # "Bye"를 출력

print("done")  # "done"을 출력
````

`elif, else`는 `if`에서 조건이 맞지 않았을 때 연속해서 조건검사를 하고 싶을 때 사용한다.


> 만약에 지갑에 12000원 이상 있으면 택시를 타고, 그게 아니고 지갑에 8000 ~ 12000원 미만 있으면 버스를 타고, 그게 아니면 걸어 간다.

````python
money = 12000

if money >= 12000:
  print("택시를 타고 집에 갑니다.")
elif money < 12000 and money >= 8000:
  print("버스를 타고 집에 갑니다.")
````

`그게 아니고 ~ 라면`이라고 시작하는 코드라면 `elif`를 쓰면 된다.

```python
예시

회원 가입이 성공하려면 글자가 영어와 숫자로만 이뤄져야 하고, 8글자 이상이어야 한다.

if not [회원 아이디가 영어와 숫자로 이루어져 있는가]:
  영어와 숫자만 쓰라고 경고창 띄워주기
if not [8글자 이상인지 검사하는 코드]:
  8글자 이상 입력하라고 경고창 띄워주기

```

````python
money = 1000

if money >= 10000:
  print("자바칩 프라프치노 먹는다.")
elif money < 10000 and money >= 5000:
  print("아메리카노 먹는다 (샷 추가).")
elif money < 5000 and money > 2500:
  print("아메리카노 먹는다.")
else:
  print("물 마시기")
````

* `if`는 독립적인 조건을 구성한다면 여러 개가 올 수있다.
* `elif`는 `if`에 대해 연속 적인 구성을 구성할 때 사용한다.
  - `0 ~ 255`개 까지 사용 가능
* `else`는 `if` 및 `elif`의 모든 조건이 맞지 않을 때 사용한다.
  - `0`개 또는 `1`개만 사용 가능


### 조건문 문제

1. ```지갑에 만원이상 있으면 택시
5000원 이상 있으면 광역버스
2000원 이상 있으면 일반버스
그게 아니면 걸어가기
```


````python
money = 10000

if money >= 5000:
  print("광역 버스")
elif money >= 2000:
  print("일반 버스")
else:
  print("걸어 간다.")
````

2. ```계좌에 10000원이 들어 있습니다.
계좌에 10000원이 들어 있습니다.
인출 금액을 입력 받습니다.
인출 금액이 계좌에 있는 금액보다 크면 "인출이 불가능합니다. {} 원의 잔액이 부족합니다." 출력
인출 금액이 계좌에 있는 금액보다 작으면 "{} 원이 인출 되었습니다." 출력
마지막에 현재 계좌의 잔액을 출력 : "현재 잔액은 {} 원 입니다." 출력
```

````python
money = int(input("얼마 인출 할래? "))
account = 10000

# 만약에 계좌에 있는 돈이 인출할 돈보다 작으면
# 인출 불가능
if account < money:
  # 인출할 돈 - 잔액
  print(f"인출이 불가능 합니다. {money - account}원의 잔액이 부족합니다.")
else:
  account -= money # account = account - money
  print(f"{money} 원이 인출 되었습니다.")

print(f"현재 잔액은 {account} 원 입니다.")
````

# 반복문( Loop )
- `while` : 조건제 반복
  - **~라면 Code Context를** 반복적으로 실행
    - ex) `exit`을 입력 할 때 까지 코드를 반복
    - ex) 상담 종료 메세지가 입력 될 때까지 반복

    - `list`, `tuple`, `set`, `dict`
  - `iterator` : 순회를 하게 하는 화살표

## `while`

```python
while [condition]:
  code context
```

````python
data = 3

# data가 0이 아니면 반복해라
while data != 0:
  print("Hello", data)
  data -= 1

print("done")
````

## 삼(3)항 연산자
- 파이썬이 아닌 다른 언어들은 삼항연산자(`?`)가 존재하지만, 파이썬에는 없음!
- `if`, `else`만 사용해서 삼항연산을 수행

간단한 `if`, `else` 구문을 **한줄의 코드**로 표현할 수 있는 방법

````python
result = None # 연산의 결과물을 저장할 변수

a = 1
b = 2

# a와 b를 더한 결과를 result에 저장
# result = a + b

if a > b:
  result = "Hello"
else:
  result = "Bye"

print(result)
````

<True 일 때 사용할 값> if [condition] else <False 일 때 사용할 값>
```

````python
result = "Hello" if a > b else "Bye"
print(result)
````

# 컴프리헨션
- `list`, `set`, `dict`를 간단하게 만들기 위한 문법
- `tuple`은 컴프리헨션으로 만들 수 없다!

반복문을 이용해서 컬렉션에 데이터를 추가하는 경우가 많다!
- 여러 전처리나 효과 없이 아주 단순하게 만들어진 데이터를 컬렉션에 추가(`append`)하고 싶으면????
- 한줄짜리 간단한 코드에 의한 아이템 추가

````python
# 1 ~ 5 까지 각각의 정수에 제곱한 결과를 리스트에 담아놓기
result = []

for data in range(1, 6):
  result.append( data ** 2 )
print(result)
````

`list` 컴프리헨션
```python
[ <추가할 값에 대한 처리> for <값> in <collection> if <값에 대한 조건> ]

<값에 대한 조건>이 True일 때 collection에 있는 아이템을 리스트에 추가
if문은 있어도 되고 없어도 된다.
```

````python
result = [ data ** 2 for data in range(1, 6) ]
result
````

````python
# 1 ~ 100 까지 홀수인 정수에 대해서 제곱한 후 추가(홀수만 뽑기)
result = [ data ** 2 for data in range(1, 101) if data % 2 == 1]
print(result)
````

````python
# 1 ~ 100 홀수는 제곱하고, 짝수는 그대로 리스트에 추가하는 컴프리헨션을 만들어 보세요
result = [ data if data % 2 == 0 else data ** 2 for data in range(1, 101) ]
print(result)
````

1. 단어의 글자가 몇 번 등장했는지에 대한 딕셔너리 만들기
    * `"letter" --> {'l' : 1, 'e' : 2, 't' : 2, 'r' : 1}`
2. 특정 문장에 단어가 몇 번 등장했는지에 대한 딕셔너리 만들기
    * `"hello hi bye hello" --> {'hello' : 2, 'hi' : 1, 'bye' : 1}`

````python
text = input("단어")
result = {i : text.count(i) for i in text}
result
````

````python
text = "hello hi bye hello"
result = {i:text.count(i) for i in text.split(" ")}
result
````

````python
text = "letter"

result = {char : text.count(char) for char in set(text)}
result
````

````python
text = "hello hi bye hello"

result = { term : text.count(term) for term in set(text.split()) }
result
````

````python
text = "letter"
letter = set(text)
number = [text.count(word) for word in letter]
result ={ letter:number for letter,number in zip(letter, number)}
print(result)
````
