- `try` : 예외가 일어날만한 코드가 들어가는 구간 (일반적인 코드)
- `except` : `try`에서 일어난 예외를 처리하는 구간 ( 예외처리 코드 )
- `finally` : `try`, `except` 구문이 실행 된 후 실행할 코드
- `raise` : 특정 상황에서 개발자가 일부러 예외를 발생 시키기 위한 코드

```python
ls = [1, 2, 3]

try:
  print("Hello")
  print(ls[3])
  print("Bye")
except:
  print("Error 입니다~!")
```

```python
try:
  print("Hello")
  print(ls[3])
  print("Bye")
except Exception as e:
  print("Error 발생!")
  print(e)

print("Done")
```

```python
try:
  print("Hello")
  print(ls[1])
  print("Bye")
except Exception as e:
  print("Error 발생!")
  print(e)
finally:
  print("Done")
```

```python
num = 10

try:
  print(num / 1)
  raise(ZeroDivisionError())
except Exception as e:
  print('Error!!!')

print("Done")
```

```python
# 나이 예외 클래스
class AgeException(Exception):

  def __init__(self, age):
    self.age = age

  def __str__(self):
    return "나이 값이 올바르지 않습니다. {}".format(self.age)
```

```python
def set_age(age):
  if age < 0:
    raise AgeException(age)

  print(age)
```

```python
set_age(-10)
```
