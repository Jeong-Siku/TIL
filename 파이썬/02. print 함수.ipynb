# print 함수
- 입력한 내용을 **출력**하는 **기능**
- 함수란? 특정 기능을 원할 때마다 실행해서 사용

```python
# print를 쓰지 않고 노트북에 출력하기
a = 1
a # 개발자용 출력
```

```python
b = 3
a
b # 제일 마지막에 위치한 변수만 출력이 된다.
```

```python
# print 를 사용해서 노트북에 출력하기
c = 1
d = 3

print(c)
print(d)
```

```python
# 쉼표로 구분하여 여러 개의 출력 하기
print(c, d)
```

```python
# sep 옵션을 이용해서 출력되는 값 사이의 문자를 지정할 수 있다.
print(c, d, sep="#")
```

출력이 `Hello 10 100 hi bye True 3.14 10` 나올 수 있도록 print 함수 사용해 보기

```python
print("Hello", 10, 100, "hi", "bye", True, 3.14, 10)
```

```python
print("Hello", 10, 100, "hi", "bye", True, 3.14, 10, sep="*")
```
