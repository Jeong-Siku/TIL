```python
from numpy.linalg import svd
import numpy as np
```

```python
# 3.4.1
B = np.array([3,2,2,2,3,-2]).reshape(2,3)
C = np.array([2,4,1,3,0,0,0,0]).reshape(4,2)
```

# 축소형X

```python
U_B, S_B, VT_B = svd(B)
```

```python
S_B_diag = np.diag(S_B,-1)[1:,:]
```

```python
U_C, S_C, VT_C = svd(C)
```

```python
S_C_diag = np.diag(S_C,2)[:,2:]
```

```python
U_B, S_B_diag, VT_B
```

```python
U_B@ S_B_diag@ VT_B
```

```python
U_C, S_C_diag, VT_C
```

```python
U_C @ S_C_diag @ VT_C
```

# 축소형

```python
B = np.array([3,2,2,2,3,-2]).reshape(2,3)
C = np.array([2,4,1,3,0,0,0,0]).reshape(4,2)
```

```python
B_U , B_S , B_VT = svd(B,full_matrices=False)
```

```python
B_U , B_S , B_VT
```

```python
B_U @ np.diag(B_S) @ B_VT
```

```python
C_U , C_S , C_VT = svd(C,full_matrices=False)
```

```python
C_U , C_S , C_VT
```

```python
C_U @ np.diag(C_S) @ C_VT
```

```python
B = np.array([3,2,2,2,3,-2]).reshape(2,3)
```

```python
U , S , VT = svd(B,full_matrices=False)
```

```python
U , S , VT
```

```python
c = B @ VT.T 
d = U @ np.diag(S)
```

```python
c, d
```

```python
B @ VT.T  == U @ np.diag(S)
```

```python
C = np.array([2,4,1,3,0,0,0,0]).reshape(4,2)
```

```python
C_U, C_S , C_V = svd(C , full_matrices=False)
```

```python
C_U, C_S , C_V
```

```python
C @ C_V.T == np.diag(C_S) @ C_U[:2,:]
```

```python
C @ C_V.T, np.diag(C_S) @ C_U[:2,:]
```
