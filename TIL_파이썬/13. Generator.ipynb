# Iterator, Generator
- `iterable` : 순회가 가능한 데이터 집합
- `iterator`
  - 반복적으로 실행 할때마다 다른 결괏값이 나오는 함수
  - 값 생성기

```python
ls = [10, 20, 30]

for i in ls:
  print(i)
```

```python
ls = [10, 20, 30]
ls_it = iter(ls)
ls_it
```

`iteration` : `iterator`에서 값을 순서대로 가져오는 행위
- `next`함수를 활용

```python
# iteration 수행
next(ls_it)
```

## 피보나치 수열
- `0 1 1 2 3 5 8 13 ...`

```python
class Fib:

  def __init__(self):
    self.prev = 0
    self.curr = 1
  
  # __next__를 구현하면 iterable한 자료형이 된다. - iterator가 될 수 있다.
  def __next__(self):
    value = self.curr

    self.curr += self.prev
    self.prev = value

    return value
```

```python
fib = Fib()
```

```python
next(fib)
```

## Generator
- `iterator`를 간단하게 만든 것
- `iter()`나 `next()`함수를 사용해서 클래스로 만들어 놓을 필요가 없다.
- `yield` 또는 `yield from` 키워드가 있으면 제너레이터가 된다.
- 일반 함수와는 달리 함수를 호출해도 함수 내부의 코드가 실행되지 않는다.

```python
def fib():
  prev, curr = 0, 1

  while True:
    yield curr
    prev, curr = curr, prev + curr
```

- `return` : 값을 반환하고 함수 자체를 끝낸다.
- `yield` : 값을 반환하고 함수를 끝내지 않고 아래에 있는 코드를 계속 실행 시킨다.

```python
f = fib()
```

```python
next(f)
```

```python
def test():
  yield 10
  yield 20
  yield 30
```

```python
t = test()
```

```python
next(t)
```

```python
def test2():
  lst = [10, 20, 30, 40, 50]
  yield from lst
```

```python
t2 = test2()
```

```python
next(t2)
```
