# 컬렉션
* 무엇인가를 모은것!
* 파이썬의 컬렉션은 **파이썬에 존재하는 모든 것을** 모을 수 있다.
  * 문법적인 요소들 빼고는 모두 다 모을 수 있다.
  * 데이터, 변수, 함수, 클래스, 오브젝트
* 어떠한 데이터를 모아낸 **구조**
  * **자료구조**



# 컬렉션의 종류
* `list`
  * `Sequence` 자료형
  * `Mutable` 자료형
    * 추가된 데이터가 바뀔 수 있다.
  * 중복 데이터 저장이 가능하다.
  * `[]` 또는  `list()` 를 이용해서 만들 수 있다.
* `tuple`
  * `Sequence` 자료형
  * `Immutable` 자료형
    * 추가된 데이터가 바뀔 수 없다.
    * `str` 형식의 데이터도 `Immutable`
  * 중복 데이터 저장이 가능하다.
  * `()`를 활용해서 만든다.
* `set`
  * `Non-Sequence` 자료형
    * 데이터 저장의 순서를 보장하지 않는다.
  * 중복 데이터의 저장도 허용하지 않는다.
  * **집합**의 개념
* **`dict`**
  * `Non-Sequence` 자료형
  * `Key-Value` 구조 자료형
* generator

## `tuple`
- 리스트와 매우 흡사하지만 변경(추가, 삭제, 수정)이 불가
- 튜플은 리스트보다 같은 데이터를 가졌을 때 공간을 적게 사용

```python
# 리스트는 아이템의 수정이 가능
lst = [1, 2, 3]
print(lst)
lst[1] = "hahahaha" # 1번 인덱스 아이템을 수정
print(lst)
```

```python
# 튜플은 아이템 수정이 불가능
tup = (1, 2, 3)
print(tup)
```

```python
tup[1] = "hahahaha"
```

```python
tup[1]
```

```python
tup[:2]
```

```python
tup[::-1]
```

```python
# tuple은 아이템 변경이 불가능한 Immutable 자료형
# str 도 Immutable
text = "abcd"
text[1] = "f" # Error
```

`"abcd"` -> `"abfcd"`

```python
text = "abcd"
text[:2]+"f"+text[2:]
```

```python
# 그냥 공부하지 마시고 참고만 하세요~
# 리스트와 튜플의 저장공간 비교
import sys

lst = [1, 2, 3]
tup = 1, 2, 3 # 소괄호 생략이 가능

print(sys.getsizeof(lst))
print(sys.getsizeof(tup))
```

## 패킹(Packing), 언패킹(UnPacking) ★★★★★
- 패킹 : 데이터를 모아내는 작업
- 언패킹 : 모아낸 데이터를 풀어 놓는 작업

```python
# 패킹
tup = ( 1, "two", 3 )

# 언패킹
item1, item2, item3 = ( 1, "two", 3 )

print(item1)
print(item3)
print(item2)
```

```python
item1, item2 = tup
```

```python
item1, item2, item3, item4 = tup
```

```python
tup = ('one', 2, 'three')

item1 = tup[0]
item2 = tup[1]
```

```python
# 관례상 필요 없는 패킹된 데이터는 언더바(_)를 변수명으로 사용
# _ 로 이름이 지어진 변수는 사용하지 않을 변수

item1, item2, _ = tup
```

```python
_
```

```python
# 1, 3만 언패킹 하고 싶을 때
tup = (1, 2, 3, 4)

item1, _, item2, _ = tup
```

수학에서의 **집합**과 동일한 개념
- 교집합 : `intersection`
- 합집합 : `union`
- 차집합 : `difference`
- 부분집합 : `issubset`

```python
A = {1, 2, 3}
B = {2, 3, 4}
```

```python
A.intersection(B) # A와 B의 교집합
```

```python
A.union(B) # A와 B의 합집합
```

```python
A.difference(B)
```

```python
# 부분집합 - True / False
C = {2, 3}
```

```python
C.issubset(A)
```

```python
C.issubset(B)
```

```python
B.issubset(A)
```

## `dict` ★★★★★
- `{key : value}`
  - `{key1 : value1, key2 : value2, key3 : value1}`
- `key`의 특징
  - `key`는 `set`의 형태로 저장된다.
  - 정수와 문자열로만 설정이 가능하다.
- `value`의 특징
  - `value`는 `list`의 형태로 저장된다.
  - 중복 저장 허용
  - 파이썬에 존재하는 모든 데이터 타입이 저장 될 수 있다.

```python
city = {
    "Korea" : "Seoul",
    "Japan" : "Tokyo",
    "Canada" : "Ottawa"
}
```

```python
print(city)
```

```python
# dict에서 데이터를 가지고 올 때는 key를 활용한다.
city["Japan"]
```

```python
city["Korea"]
```

```python
city["usa"]
```

```python
city.get("usa", "없음")
```

### `dict`에 원소 추가 및 변경
- 추가와 변경의 문법이 똑같다.
- 키가 없으면 추가, 있으면 수정

```python
city["China"] = "Beijing"
city
```

```python
city["Japan"] = "Kyoto"
city
```

### `dict` 원소 삭제
- `del` 키워드를 사용

```python
del city["Japan"] # city 딕셔너리에서 키가 Japan인 원소를 삭제
city
```

### `dict`의 `key`, `value`를 확인 ★★★★★

```python
# keys() : 딕셔너리 내의 전체 키를 확인
city.keys()
```

```python
# values() : 딕셔너리 내의 전체 value를 확인
city.values()
```

```python
# items() : 딕셔너리 내의 전체 key, value를 튜플 형태로 확인
city.items()
```

```python
for i in city.items():
    print(i)
```

```python
a = {
    "a" : 1,
    "b" : 1,
}

a.values()
```

```python
a = {
    "a" : 1,
    "b" : 1,
    "b" : 3
}

a
```
