```python
import re
```

```python
# 검사 대상 문자열 : 소스
source = "I can do it! You can do it!"
pattern = "You"

# patten에 맞는 문자열로 시작하느냐?
m = re.match(pattern, source)
print(m)
```

```python
# 검사 대상 문자열 : 소스
source = "I can do it! You can do it!"
pattern = "You"

# patten에 맞는 문자열로 시작하느냐?
m = re.search(pattern, source)
print(m)
```

```python
source = "이런 내맘 모르고 너무해 너무해"

pattern = "너.*"
re.search(pattern, source).group()
```

```python
pattern = "너무해"
re.findall(pattern, source)
```

```python
twice_tt = """
I'm likes TT
Just liae TA
이런 내 맘 모르고 너무해 너무해
I'm lidd TB
Just licc TC
Tell me that you'd be my baby
"""

re.split("T.?", twice_tt)
```

```python
print(re.sub("li.{2}", "love", twice_tt))
```

```python
pattern = "ca*t" # c로 시작하면서 a가 0번이상 반복되고 끝에 t

re.findall(pattern, "ct")
```

```python
re.findall(pattern, "caaaat")
```

```python
pattern = "ca+t" # a가 1번 이상 등장
 
re.findall(pattern, "ct")
```

```python
re.findall(pattern, "caaat")
```
