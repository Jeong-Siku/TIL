# 연산자의 종류
- 계산(연산)을 할 수 있는 **기호**

## 산술연산자
- `+, -, *, /`
- `//, %`

```python
print( 5 + 2 )
print( 5 - 2 )
print( 5 * 3 )
print( 5 / 2 )
```

```python
print( 5 // 2 ) # 5 나누기 2의 몫 구하기
print( 5 % 2 ) # 5 나누기 2의 나머지 구하기
```

```python
print( 2 + 2 * 2 ) # 곱하기가 우선적으로 수행된다.
```

```python
# 소괄호를 이용해서 연산의 우선순위를 조절할 수 있다.
print( (2 + 2) * 2 )
```

## 할당(대입) 연산자
- 어떤 변수에 값을 대입하는 연산자
- 특정 변수에 들어있는 값에 **특정 연산**을 수행해서 그 변수에 다시 저장(누적)
- `=, +=, -=, ...`

```python
num = 10
print(num)
```

```python
# 누적이란? 값이 계속해서 쌓여가는 것
num = num + 5
print(num)
```

```python
num += 5 # num = num + 5
print(num)
```

```python
num /= 5
print(num)
```

## 비교연산자
- 값이나 변수 끼리 크기를 비교
- `>, <, >=, <=`
- `!=, ==`

비교연산자의 결과물은 무조건 **True / False**

```python
a = 5
b = 3

print(a > b) # True
print(a < b) # False
print(a >= 5) # True
print(b <= 3) # True
```

```python
# != : 부정연산자, == : 동일연산자
print( a == b ) # a와 b가 같다 -> False
print( a != b ) # a와 b가 같지 않다. -> True
```

## 논리연산자
- 논리 자료형인 `bool`을 연산하는 것
- `and, or, not`

**and( & )**
* `A and B` : A가 `True`, B도 `True`면 결과물이 `True`

**or( | )**
* `A or B` : A와 B 둘 중 하나만 `True`여도 결과물이 `True`

**not( ! )**
* `not A` : A가 `True`면 `False`가 되고, `False`면 `True`가 된다.


```python
1 < 3 and 3 > 1
```

```python
1 > 3 and 3 > 1
```

**`and` 연산의 원리**

`A and B`
* `A`가 True면 `B`를 연산을 한다.
* `A`가 False면 `B`를 연산하지 않는다.


```python
1 > 3 or 3 > 1
```

```python
3 < 1 or 4 < 3
```

**`or` 연산의 원리**

`A or B`
* `A`가 True면 `B`를 연산하지 않는다.
* `A`가 False면 `B`를 연산을 한다.


```python
not 1 > 3 # 1이 3보다 크지 않다.
```

```python
not 1 < 3 # 1이 3보다 작지 않다.
```

## 멤버연산자
- 멤버란? 컬렉션 안에 있는 요소, 원소, item
- 컬렉션에 원소가 존재하는지 여부를 판단
- `in, not in` 

```python
lst = ["a", 1, 3.1]

"a" in lst
```

```python
2 in lst
```

```python
2 not in lst
```

```python
"h" in "hello"
```

```python
temp_dict = {
    "a": 10,
    "b": 20
}

"a" in temp_dict # dict의 in은 key를 비교한다.
```

```python
10 in temp_dict.values() # value가 들어 있는지 확인하기
```

```python
lst = [1, 1, 2, 2, 3, 3, 3]
lst.count(1) # 멤버(item)의 개수 세기
```

```python
"hello".count("l")
```
