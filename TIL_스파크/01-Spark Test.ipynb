```python
from pyspark import SparkConf, SparkContext
```

```python
# 스파크 환경 설정 객체 생성
# 마스터를 만들어 놓고 시작해야한다.
conf = SparkConf().setMaster("local").setAppName("country-student-counts") 
# hadoop없이 설정하기에 local로 설정

# SparkContext 생성
sc = SparkContext(conf=conf) # RDD만 쓴 환경
# sc라는 변수명은 변경하지말고 그대로 쓸 것!!
# submit으로 배포할 때 sc가 필요
```

```python
filepath="/home/ubuntu/working/spark-examples/data/xAPI-Edu-Data.csv"
```

```python
# 로컬 파일에서 데이터를 불러올 것이기 때문에 스킴이 file://
# 만약 HDFS에서 불러오려면? hdfs://

lines = sc.textFile(f"file:///{filepath}")
lines
# RDD 형성
# 추상화된다. 하나의 파일처럼 lines에
```

```python
header = lines.first()
header
```

```python
# 헤더를 제외한 데이터만 불러오기
datas = lines.filter(lambda row: row != header)# Filter : Transformations
datas
```

```python
# MAP을 해서 데이터를 쪼개기
countries = datas.map(lambda row : row.split(",")[2])
countries
# Transformation
```

```python
# RDD에 있는 내용 보기
countries.collect() # 액션. 모든 Transformations 작업을 한꺼번에 수행
```

```python
# 각 나라별 개수 확인
result = countries.countByValue()
```

```python
# 작업이 끝났으면 반드시 스파크 컨텍스트를 종료
sc.stop()
```
